```python
import dendropy
import pandas as pd

import phylo
```

```python
records_pseudo = pd.read_csv('../data/pseudo/responses-pseudo.csv', names=['id', 'lab-id', 'carb', 'toby'], skiprows=1)
records_staph = pd.read_csv('../data/staph/responses-staph.csv', names=['resp', 'id'], skiprows=1)
```

```python
records_pseudo.set_index('id', inplace=True)
records_staph.set_index('id', inplace=True)
label_pseudo = records_pseudo['carb'].dropna()
label_staph = records_staph['resp'].dropna()
```

```python
phylo.draw_tree('../../zhaoyi/mega/pseudo/pseudo_ml.nwk', 'ml_pseudo',
                '$Pseudomonas$ maximum likelihood phylogenetic tree',
                labels=label_pseudo)
```

```python
phylo.draw_tree('../../zhaoyi/mega/staph/staph_ml.nwk', 'ml_staph',
                '$Staphylococcus$ maximum likelihood phylogenetic tree',
                label_staph)
```
